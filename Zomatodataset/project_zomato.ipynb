```python
df_country=pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
df_country.columns
```

```python
df_country.isnull().sum()
```

```python
df.columns
```

```python
#notice how country code is a common attribute among the two data frames hence we think of combining them 
```

```python

pd.merge(df,df_country,on='Country Code',how='left') #merges two df, on soecifies common attribute, left specifies left join 
```

```python
final_df=pd.merge(df,df_country,on='Country Code',how='left')
```

```python
final_df.columns
```

```python
#notice how the final_df has attribute of both the dfs
```

```python
#another way to check data types 
final_df.dtypes
```

```python
final_df.Country.value_counts()
#gives count of values for each country
#we can conclude that India will have the most transactions here 
```

```python
country_names=final_df.Country.value_counts().index
#lists the names in the order shown above 
```

```python
country_names
```

```python
#to plot a pie chart we need array of values 
#creating array
final_df.Country.value_counts().values
```

```python
y=final_df.Country.value_counts().values #storing array in y
```

```python
plt.pie(y,labels=country_names)  #plotting against y 
plt.show()  #to show the plot 
```

```python
#notice how the plot is cluttered. instead we can chose to show the top three values only
import matplotlib
matplotlib.rcParams['figure.figsize']=(18,6)
plt.pie(y[:3],labels=country_names[:3],autopct='%.2f')  #autopct='%.2f' is used to show the % in the pie chart 
plt.show()
```

```python
#obs 1: the top 3 countries with max values/transactions are - India, US, UK.
```

```python
#we notice that several clms have rating associated with it so we group them together 
final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size()
```

```python
#obs 2:from here we can observe how with rating, the colour and rating text change 
```

```python
final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

```python
ratings=final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

```python
ratings
```

```python
#we plot a bar graph to visualise rating data 
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(18,6)
sns.barplot(x="Aggregate rating",y="Rating count",data=ratings)
```

```python
#to get the colour of the bars to be the colour of the rating, put in parameter of hue 
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(18,6)
sns.barplot(x="Aggregate rating",y="Rating count",hue='Rating color',data=ratings,palette=['white','red','orange','yellow','green','green'])
```

```python
#the palette sets the colour in order of the colour rating 
#palette can also be used to change the colour of the bars 
```

```python
sns.countplot(x='Rating color', data=ratings,palette=['white','red','orange','yellow','green','green'])
```
